```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cycles = 50

N = np.genfromtxt('usual_matrix_multiplication.txt')[:cycles,0]
usual_matmul = np.genfromtxt('usual_matrix_multiplication.txt')[:cycles,1]
cache_matmul = np.genfromtxt('cache-friendly_matrix_multiplication.txt')[:cycles,1]
fortran_matmul = np.genfromtxt('fortran_matrix_multiplication.txt')[:cycles,1]

usual_matmul_optimized = np.genfromtxt('usual_matrix_multiplication.txt')[cycles:2*cycles,1]
cache_matmul_optimized = np.genfromtxt('cache-friendly_matrix_multiplication.txt')[cycles:2*cycles,1]
fortran_matmul_optimized = np.genfromtxt('fortran_matrix_multiplication.txt')[cycles:2*cycles,1]
```

```python
MEDIUM_SIZE = 16
plt.rc('axes', labelsize=MEDIUM_SIZE)

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(N,usual_matmul, label= 'usual')
ax.plot(N,cache_matmul, label = 'cache friendly')
ax.plot(N, fortran_matmul, label = 'Fortran matmul')
ax.set(xlabel = 'Square Matrix Size N', ylabel = 'CPU elapsed time', title = 'Matrix multiplication, optimization flag = OFF')
ax.legend()
ax.grid(True)

plt.savefig('matmul_plot.pdf')
```

```python
fig_opt, ax_opt = plt.subplots(figsize=(10,10))

ax_opt.plot(N,usual_matmul_optimized, label= 'usual')
ax_opt.plot(N,cache_matmul_optimized, label = 'cache friendly')
ax_opt.plot(N, fortran_matmul_optimized, label = 'Fortran matmul')
ax_opt.set(xlabel = 'Square Matrix Size N', ylabel = 'CPU elapsed time', title = 'Matrix multiplication, optimization flag = O3')
ax_opt.legend()
ax_opt.grid(True)

plt.savefig('matmul_plot_optimized.pdf')
```
