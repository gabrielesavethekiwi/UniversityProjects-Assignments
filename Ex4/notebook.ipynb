```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.hermite import *
from matplotlib.pyplot import figure
import matplotlib.ticker as ticker

SMALL_SIZE = 10
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def th_eigenfunction(N, xs):
# xs is the 1D grid of points, the lattice.
    m = 1
    h_bar = 1
    omega = 1


    # Hermite polynomial, like ax^n+bx^{n-1}+... We only want the last term
    herm_coeff = []
    for i in range(N):
        herm_coeff.append(0)
    herm_coeff.append(1)

    normalization = 1 / ((2**N) * np.math.factorial(N))**0.5 * (m*omega/(np.pi*h_bar))**0.25
    #psi = np.exp(-m*omega*xs**2/(2*h_bar)) * hermval((m*omega/h_bar)**0.5 * xs, herm_coeff)
    #numpy.polynomial.hermite.hermval
    psi = np.exp(-m*omega*xs**2/(2*h_bar)) * np.polynomial.hermite.hermval((m*omega/h_bar)**0.5 * xs, herm_coeff)
    
    psi = np.multiply(psi, normalization)
    return psi

def th_eigenvalues(N):
    h_bar = 1
    omega = 1
    
    E = h_bar*omega*(N+0.5)
    return E
```

```python
i = 5
a = 5
fname = 'eigenvalues'+str(i)+'00_'+str(a)+'_1.txt'
eigenvalues = np.genfromtxt(fname)

fname_functions = 'eigenfunctions'+str(i)+'00_'+str(a)+'_1.txt'
eigenfunctions = np.genfromtxt(fname_functions)

#dimension = 2*N+1 where N is the input to the fortran program
dimension = len(eigenvalues)
N = (dimension-1)/2
dx = a / N

grid = np.arange(-N, N+1)*dx

#plt.plot(grid, eigenfunctions) you can plot ALL of the columns together
#very beautiful

#the eigenfunction that you want to select
index = 0

figure(figsize=(16, 9))
plt.plot(grid, eigenfunctions[:,index], label = 'N ='+str(i)+'00')
plt.plot(grid, th_eigenfunction(index, grid), label='theoretical')
plt.title('Eigenfunction #'+ str(index) + ';   dx = ' + str(dx))
plt.xlabel('x')
plt.legend()
```
